## Exercise 1: Plotting with ForceAtlas2

```python
import networkx as nx
import random as rd
from IPython.display import Image, display

# Nodes and average degree
N = 200
deg_avg = 1.5

# Calculate p to achieve average degree k=degree_avg
p = deg_avg / (N - 1)

# Erdos-Renyi (ER) network
ER_graph = nx.erdos_renyi_graph(N, p)

# Plotting the whole network
# nx.draw(ER_graph, node_size=70, node_color="#a9cae0")

# Extract connected components
components = list(nx.connected_components(ER_graph))
# print(components)

# Find the largest component (GCC)
gcc_nodes = max(components, key=len)
GCC = ER_graph.subgraph(gcc_nodes).copy()

# Plotting GCC
nx.draw(GCC, node_size=70, node_color="#a9cae0", edgecolors="k")

```
